## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns

y_df = attrition_df[['Attrition','Department']]
y_df
```

```python
# Create a list of at least 10 column names to use as X data
X_data = ['Education',
          'Age',
          'DistanceFromHome',
          'JobSatisfaction',
          'OverTime',
          'StockOptionLevel',
          'WorkLifeBalance',
          'YearsAtCompany',
          'YearsSinceLastPromotion',
          'NumCompaniesWorked']

# Create X_df using your selected columns
X_df = attrition_df[X_data]


# Show the data types for X_df
X_df.dtypes
```

```python
# Split the data into training and testing sets
#from sklearn.model_selection import train_test_split

X_df_train, X_df_test, y_df_train, y_df_test = train_test_split(X_df, y_df, random_state=42)
```

```python
# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary

X_df['OverTime'] = X_df['OverTime'].apply(lambda x: 1 if x == 'yes' else 0)
X_df
```

```python
## Create a StandardScaler
Scaler = StandardScaler()

# Fit the StandardScaler to the training data
Scaler.fit(X_df_train)

# Scale the training and testing data
X_df_train_scaled = Scaler.transform(X_df_train)
X_df_test_scaled = Scaler.transform(X_df_test)
X_df_train_scaled
```

```python
# Create a OneHotEncoder for the Department column
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
department_encoder = OneHotEncoder(sparse_output=False)

# Fit the encoder to the training data
dept_encoder = department_encoder.fit(y_df_train[['Department']])

# Create two new variables by applying the encoder
# to the training and testing data
dept_train_encoded = dept_encoder.transform(y_df_train[['Department']])
dept_test_encoded = dept_encoder.transform(y_df_test[['Department']])
dept_test_encoded
```

```python
# Create a OneHotEncoder for the Attrition column
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
attrition_encoder = OneHotEncoder(sparse_output=False)

# Fit the encoder to the training data
attr_encoder = attrition_encoder.fit(y_df_train[['Attrition']])

# Create two new variables by applying the encoder
# to the training and testing data
attr_train_encoded = attr_encoder.transform(y_df_train[['Attrition']])
attr_test_encoded = attr_encoder.transform(y_df_test[['Attrition']])
attr_test_encoded
```

## Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data
num_columns = len(X_df_train.columns)

# Create the input layer
input_layer = layers.Input(num_columns)

# Create at least two shared layers
shared_layer1 = layers.Dense(64, activation='relu')(input_layer)
shared_layer2 = layers.Dense(128, activation='relu')(shared_layer1)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer

# Create the hidden layer
department_hidden = layers.Dense(32, activation='relu')(shared_layer2)

# Create the output layer
Department_output = layers.Dense(3, activation='softmax', name='Department_output')(department_hidden)
```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer

# Create the hidden layer
attrition_hidden = layers.Dense(32, activation='relu')(shared_layer2)

# Create the output layer
Attrition_output = layers.Dense(2, activation='sigmoid', name='Attrition_output')(attrition_hidden)
```

```python
# Create the model
model = Model(inputs=input_layer, outputs=[Department_output, Attrition_output])

# Compile the model
model.compile(optimizer='adam',
              loss={'Department_output': 'categorical_crossentropy', 'Attrition_output': 'binary_crossentropy'},
              metrics={'Department_output': 'accuracy', 'Attrition_output': 'accuracy'})

# Summarize the model
model.summary()
```

```python
# Train the model
batch_size = 32
epochs = 100

model.fit(
    X_df_train_scaled,
    {
        'Department_output': dept_train_encoded,
        'Attrition_output': attr_train_encoded,
    },
    epochs=epochs,
    batch_size=batch_size,
    validation_split=0.2
)
```

```python
# Evaluate the model with the testing data
test_results = model.evaluate(np.array(X_df_test_scaled), {
        'Department_output': dept_test_encoded,
        'Attrition_output': attr_test_encoded,
    })
test_results
```

```python
# Print the accuracy for both department and attrition
print(f"Department predictions accuracy: {test_results[3]}")
print(f"Attrition predictions accuracy: {test_results[4]}")
```
